```python
import cv2
import mediapipe as mp 
import numpy as np
```

```python
mp_face_mesh = mp.solutions.face_mesh
```

```python
img = cv2.imread("./test.jpeg")

with mp_face_mesh.FaceMesh(
    static_image_mode=True,
    max_num_faces=1,
    refine_landmarks=True,
    min_detection_confidence=0.5) as face_mesh:
    # Convert the BGR image to RGB before processing.
    results = face_mesh.process(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    # Print and draw face mesh landmarks on the image.
    if results.multi_face_landmarks:
        print(results.multi_face_landmarks[0])
```

```python
face_detector = mp_face_mesh.FaceMesh(static_image_mode=True,
    max_num_faces=1,
    refine_landmarks=True,
    min_detection_confidence=0.5)


img = cv2.imread("./test.jpeg")
results = face_detector.process(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
np_landmark = np.array([[landmark.x, landmark.y, landmark.z] for landmark in list(results.multi_face_landmarks[0].landmark)])
```
